```python
#import
import pandas as pd

#Read CSV file
df = pd.read_csv("vine_table.csv")
count_df = len(df)
df
```

```python
#filter out reviews with total votes thar are greater or equal to 20
filtered_df = df.loc[df['total_votes']>=20]
count_filtered = len(filtered_df)
filtered_df
```

```python
# Final filter with items helpful votes out of total votes are greater or equal to 50%
clean_df= filtered_df.loc[filtered_df['helpful_votes']/filtered_df['total_votes']>=0.5]
count_clean = len(clean_df) 
clean_df
```

```python
# Filter out all Vine users reviews
vineuser_df = clean_df[clean_df['vine'] == 'Y']
count_vineusers = len(vineuser_df)
vineuser_df
```

```python
#FFilter out all non vine users reviews
vinenon_df = clean_df[clean_df['vine'] == 'N']
count_vinenon = len(vinenon_df)
vinenon_df
```

```python
# Number of 5 star reviews with N as Vine
count_five_star_non = len(vinenon_df[vinenon_df['star_rating'] == 5])
count_five_star_non
```

```python
# Number of 5 star reviews with Y as Vine
count_five_star_users = len(vineuser_df[vineuser_df['star_rating'] == 5])
count_five_star_users
```

```python
# Percentage of 5 star reviews for Vine Reviews
per_of_fivestar_user = count_five_star_users / count_vineusers
per_of_fivestar_user
```

```python
# Percentage of 5 star reviews Non-Vine Reviews
per_of_fivestar_non = count_five_star_non / count_vinenon
per_of_fivestar_non
```

```python
# Printing all results
print(f"Total number of reviews returned from database vine table : {count_df:,}")
print(f"Total number of reveiws with total votes are greater or equal to 20 : {count_filtered:,}")
print(f'Total number of reviews with total helpful votes is equal or greater than 50% : {count_clean:,}')
print(f'Total number of Vine reviews: {count_vineusers:,}')
print(f'Total number of non-Vine reviews : {count_vinenon:,}')
print(f'Total number of Vine reviews with five star reviews : {count_five_star_users:,}')
print(f'Total number of non-Vine reviews with five star reviews : {count_five_star_non:,}')
print(f'Percentage of five star reviews in Vine reviews : {per_of_fivestar_user: .1%}')
print(f'Percentage of five star reviews in non-vine reviews : {per_of_fivestar_non: .1%}')
```
